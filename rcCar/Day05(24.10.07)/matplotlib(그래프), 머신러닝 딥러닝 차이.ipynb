```python
points = [np.deg2rad(90),    # 라디안: 호의길이와 반지름의 길이가 같으면 1라디안
         np.deg2rad(18),     # 1라디안: 57.3도
         np.deg2rad(306),
         np.deg2rad(234),
         np.deg2rad(162)]
x = [np.cos(points[0]),
    np.cos(points[3]),
    np.cos(points[1]),
    np.cos(points[4]),
    np.cos(points[2]),
    np.cos(points[0])]
y = [np.sin(points[0]),
    np.sin(points[3]),
    np.sin(points[1]),
    np.sin(points[4]),
    np.sin(points[2]),
    np.sin(points[0])]

plt.ylim(-1, 1)    # 좌표범위
plt.xlim(-1, 1)
plt.rcParams["figure.figsize"] = (5, 5)    # pyplot 사이즈
plt.plot(x, y)    # 선형그래프
plt.show()
```
